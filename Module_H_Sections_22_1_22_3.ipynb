**Section 22.1 ODE Initial Value Problem Statement**

* A differential equation is a felationship between a function, its independent variable and any number of its derivatives
* An Ordinary Differential Equation (ODE) is a form of differential equation where the independent variable is in one dimension
* A partial differential equation (PDE) is a general form differential equation where x is a vector containing the independent variables $x_1$ $x_2$,$x_3$,â€¦,$x_m$ ,and the partial derivatives can be of any order and with respect to any combination of variables.

**Section 22.2 Reduction of Order**

* Many numerical methods for solving IVP's are specifically designed for first-order differential equations, therefore we must often reduce the order down to first order
* To reduce the order of a differential equation one must take the derivative of the state vector S(t)

**Section 22.3 The Euler Method**

* Explicit Euler Formula allows us to compute an approximation of for the state at S($t_j+1$) given the state at S($t_j$)
* Euler Formula is the simplest and most intuitive method for solving IVP's

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

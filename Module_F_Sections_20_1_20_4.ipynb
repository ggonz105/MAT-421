**Section 20.1 Numerical Differentiation**

* Continuous functions may be discretized for numerical calculations that corresponds to data sets that may be acquired in engineering and science practice
* A derivative of a discretized function can be taken, which results in an approximated derivative

**Section 20.2 Finite Difference Approximating Derivatives**

* Finite difference approximations of a slope, x = a, can be taken using values of the function in the neighborhood of the point x = a
* Forward Difference Method: $f'(x_j)=\frac{f(x_j+1)-f(x_j)}{(x_j+1)-(x_j)}$


* Backward Diffeence Method: $f'(x_j)=\frac{f(x_j)-f(x_j-1)}{(x_j)-(x_j-1)}$


* Central Difference Method: $f'(x_j)=\frac{f(x_j+1)-f(x_j-1)}{(x_j+1)-(x_j-1)}$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Step size
h = 0.1

# Define grid
x = np.arange(0, 2*np.pi, h)
y = x**3 + 7*x
forward_diff = np.diff(y)/h # Forward difference approximation
x_diff = x[:-1:]

exact_solution = 3*(x_diff)**2 + 7 # Compute exact solution

plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
label = 'Exact solution')
plt.legend()
plt.show()

max_error = max(abs(exact_solution - forward_diff)) # Max error between numerical derivative and exact solution
print(max_error)
```

**Section 20.4 Numerical Differentiation with Noise**

* Sometimes data can be contaminated with noise, which causes the value to be off by a small amount from the pure mathematical function
* Noise can often occur in engineering from inaccuracies in measurement devices

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 2*np.pi, 0.01)
# compute function
omega = 100
epsilon = 0.01

y = -np.sin(x)
y_noise = y + epsilon*omega*np.cos(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'Derivative cos(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'Derivative of cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

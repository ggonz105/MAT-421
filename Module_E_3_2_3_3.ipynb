**Section 3.2 Continuity and Differentiation**

* A limit is the value that a given function approaches at a certain value x = $x_0$
* A function is deemed continuous if there are no abrupt changes in value (also known as discontinuities)
* The rate at which a function changes is known as the derivative of the function, which is given by the equation $f'(x_0)=lim_{h->0}\frac{f(x_0+h)-f(x_0)}{h}$
* Taylor's Theorem gives an approximation of a differentiable function around any given point by a polynomial

```python
# Derivative Example
import sympy as sp

x = sp.Symbol('x')
h = sp.Symbol('h')

f = x**3 + 2*x**2 + 3*x + 4

limit_expression = (f.subs(x, x + h) - f) / h
derivative_f = sp.limit(limit_expression, h, 0)
print(derivative_f)
```

**Section 3.3 Unconstrained Optmization**

* Local maxima and minima exist at the critical points, which are where the derivative of a function = 0
* In order to determine if the critical point results in a maxima or minima, take a derivative to the left and right of the 0. If the sign changes from (-) to (+) then it's a minima, and vice versa for a maxima

```python
# Local and Global Minimum's Example

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def f(x):
    return x**4 - 6*x**2 + x

x_values = np.linspace(-2, 2, 500)
y_values = f(x_values)

plt.plot(x_values, y_values, label='f(x) = x^4 - 4x^2 + x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the function')
plt.grid(True)
plt.legend()

# Find global minimum using optimization
result_global = minimize(f, x0=0)
x_global_min = result_global.x
y_global_min = result_global.fun

# Plot global minimum
plt.plot(x_global_min, y_global_min, 'ro', label=f'Global Minimizer: ({x_global_min[0]:.2f}, {y_global_min:.2f})')

# Find local minimum using optimization (starting from x=2)
result_local = minimize(f, x0=2)
x_local_min = result_local.x
y_local_min = result_local.fun

# Plot local minimum
plt.plot(x_local_min, y_local_min, 'go', label=f'Local Minimizer: ({x_local_min[0]:.2f}, {y_local_min:.2f})')
plt.legend()
plt.show()
```

**Section 17.1 Interpolation Problem Statement**

Assume there is a data set with independent values of x and y, and there is a desired value that lies within those data sets but is not explicitly defined. This value can then be estimated through interpolation of the data sets.

**Section 17.2 Linear Interpolation**

Linear interpolation is the simplest form of interpolating data. When using linear interpolation, it is assumed that the point to be found lies on the line joining the nearest points to the left and right.

This results in the following generalized linear interpolation equation:
y(x)=yi+(yi+1−yi)(x−xi)/(xi+1−xi), where  xi  is the nearest point less than x and xi+1 is the nearest point greater than x, and (yi+1−yi)/(xi+1−xi) is the slope and yi is the y value that corresponds to the x value of xi.

```python
import numpy as np
import matplotlib.pyplot as plt

x = [4, 5, 6]
y = [3, 10, 7]

x_interp = 5.3
y_interp = round(np.interp(x_interp, x, y),1)
print(y_interp)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(5.3, y_interp, 'ro')
plt.title('Linear Interpolation at x = 5.3')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
